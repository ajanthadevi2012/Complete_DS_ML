# Sparse DataFrames

```python
arr = np.random.randn(10000)
arr[arr < 1] = np.nan
df = pd.DataFrame(arr.reshape(100,100))
df
```

```python
'{:0.2f} bytes'.format(df.memory_usage().sum() / 1e3)
```

```python
sdf = df.astype(pd.SparseDtype("float", np.nan))
```

```python
sdf.sparse.density
```

```python
'{:0.2f} bytes'.format(sdf.memory_usage().sum() / 1e3)
```
