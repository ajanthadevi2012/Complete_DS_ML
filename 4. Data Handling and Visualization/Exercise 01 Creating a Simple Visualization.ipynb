## Exercise1: Creating a Simple Visualization

In this exercise, we will create our first simple plot using Matplotlib.

#### Import statements

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Creating, plotting, and saving a figure

```python
plt.figure(dpi=100)
# [x axis] [ y axis]
#(x,y)
plt.plot([1, 2, 4, 5], [1, 3, 4, 3], '-o')
# plt.savefig has to be called before plt.show, since plt.show closes the figure
plt.savefig('Exercise-01.png', dpi=300, bbox_inches='tight') 

plt.show()
```
